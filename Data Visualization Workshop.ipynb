```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.linspace(0,20, num = 100)
```

```python
# Generate some data
x = np.array(np.linspace(-2,2, num = 50))
y = x ** 2
y_3 = x ** 3
```

```python
# plot data using Matplotlib
plt.plot(x,y, 'b*')#'o', color = 'blue'
plt.plot(x, y_3, 'go')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('compare curves', fontsize = 20, color = 'r')
plt.legend(['order = 2', 'order = 3']) # , loc = 'lower right'
```

```python
# plot data using scatter plot
sns.scatterplot(x=x, y=y)
sns.scatterplot(x=x, y=y_3, marker="+", color = 'black')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('compare curves', fontsize = 20, color = 'r')
plt.legend(['order = 2', 'order = 3'])
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x, y_3)
```

```python
# Create two subplots using Matplotlib
fig, axis = plt.subplots(1, 2, figsize=(12, 4))
axis[0].scatter(x, y)
axis[0].set_title("order = 2")
axis[0].set_xlabel("X Values")
axis[0].set_ylabel("Y Values")

axis[1].scatter(x, y_3)
axis[1].set_title("order = 3")
axis[1].set_xlabel("X Values")
axis[1].set_ylabel("Y Values")
```

```python
# Create two subplots using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the degree 2 curve using Matplotlib on the first subplot
ax1.scatter(x, y)
ax1.set_title("Matplotlib Scatter Plot")
ax1.set_xlabel("X Values")
ax1.set_ylabel("Y Values")

# Create a scatter plot using Seaborn on the second subplot
sns.scatterplot(x=x, y=y, ax=ax2)
ax2.set_title("Seaborn Scatter Plot")
ax2.set_xlabel("X Values")
ax2.set_ylabel("Y Values")

# Display the plots
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

order = 1

for ax in axes.flat :
    yy = x ** order
    ax.plot(x, yy)
    ax.set_title(f"order = {order}")
    ax.set_xlabel("X Values")
    ax.set_ylabel("Y Values")
    order += 1
```

# Dataset : Waiter's Tips Dataset

### dataset link : https://www.kaggle.com/datasets/aminizahra/tips-dataset

### One waiter recorded information about each tip he received over a period of a few months working in one restaurant. In all he recorded 244 tips.

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
sns.histplot(data=df, x = 'tip', bins= 20, kde= True)
```

```python
sns.catplot(data=df, x = 'smoker', y = 'tip')
```

```python
sns.boxplot(data = df, x = 'tip' , y = 'day')
```

```python
sns.barplot(data = df, x = 'tip' , y = 'day')
```

```python
# categorical columns vs tips

fig, axes = plt.subplots(1, 3 , figsize = (15,4))
categorical_column = 'day'
sns.stripplot(data=df, x = categorical_column, y = 'tip' , ax = axes[0]) # catplot
sns.boxplot(data = df, x = categorical_column , y = 'tip', ax = axes[1])
sns.barplot(data = df, x = categorical_column , y = 'tip', ax = axes[2])
```

```python
import ipywidgets
```

```python
categorical_column_list = ['sex','smoker','day','time','size']

def my_plot (categorical_column = 'day', a = 10):
    fig, axes = plt.subplots(1, 3 , figsize = (15,4))
    sns.stripplot(data=df, x = categorical_column, y = 'tip' , ax = axes[0]) # catplot
    sns.boxplot(data = df, x = categorical_column , y = 'tip', ax = axes[1])
    sns.barplot(data = df, x = categorical_column , y = 'tip', ax = axes[2])
    fig.suptitle("my plot", fontsize= a)
```

```python
ipywidgets.interact(my_plot, categorical_column = categorical_column_list , a = (10, 50, 1))
```

```python
# total_bill vs tips
my_pallet = {'Yes' : 'red', 'No'  : 'black'}
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'smoker', palette = my_pallet)
```

```python
# rel plot
sns.relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'smoker') # row
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter',
            col = 'smoker', row = 'time') # row
```

```python
## size, hue , style in rel plot
sns.relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter', 
            col = 'smoker', row = 'time', 
            col_order= ['Yes', 'No'], size = 'size') # row
```

```python
# joint plot
sns.jointplot (data = df, x = 'total_bill', y = 'tip',  palette='flare', )
```

# Dataset : USA Sales Product Dataset

```python
df = pd.read_csv('Sales_Product_Combined.csv')
df.head()
```

```python
# create pie chart
pie_chart_data = df['City'].value_counts()
pie_chart_data
```

```python
explode = (0.1, 0, 0, 0, 0, 0, 0, 0.2, 0.3)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
colors = ['orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'teal']

plt.figure(figsize = (10,10))
plt.pie(pie_chart_data.values, labels=pie_chart_data.index, explode = explode, autopct='%1.2f%%', colors = colors)
```

```python
sns.countplot(data = df , x = 'Product Type')
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (12,4))
sns.countplot(data = df , x = 'Product Type', hue = 'City')
plt.xticks(rotation = 45)
```
